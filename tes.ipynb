```python
import pandas as pd 
import numpy as np 

imp = pd.read_excel('kpi_data_clean.xlsx')
```

```python
var = pd.read_csv('kpi_variables.csv')
```

```python
br = pd.read_csv('branches.csv')
```

```python
imp
```

```python
imp = imp.merge(
    br[['branch_id', 'KDO_BSI']],
    left_on='KDO_BSI',
    right_on='KDO_BSI',
    how='left'
)
imp = imp.merge(
    var[['variable_id', 'var_code']],
    left_on='var_code',
    right_on='var_code',
    how='left'
)
imp
```

```python
target = imp[['branch_id', 'variable_id', 'Periode','target_value']]
target
```

```python
realization = imp[['branch_id', 'variable_id', 'Periode','realization_value']]
realization
```

```python
var_calc = target.merge(realization, on=['branch_id', 'variable_id', 'Periode'])
var_calc
```

```python
var_calc = var_calc.merge(
    var[['variable_id','var_code','weight','type','category_id']],
    left_on='variable_id',
    right_on='variable_id'
)
```

```python
var_calc['pencapaian'] = var_calc['realization_value'] / var_calc['target_value']
```

```python
var_calc['max_target'] = var_calc['pencapaian'].clip(upper=1.5)
```

```python
var_calc['score'] = var_calc['weight'] * var_calc['max_target'] * 100
```
